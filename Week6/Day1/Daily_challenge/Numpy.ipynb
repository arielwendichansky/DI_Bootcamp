```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Data preparation
Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns

```python
temperature = np.random.uniform(-5,35,size=(10,12))
temperature
```

```python
count = 0
cities = 1
cities_data = {}
while count <= 9:
    city_data = temperature[count]
    city_data = np.expand_dims(city_data, axis=0)  # Adding new axis that will be the Index
    city_df = pd.DataFrame(city_data, index=[1], columns=range(1, 13))
    cities_data['city_{}'.format(cities)] = city_df
    count += 1
    cities += 1

print(cities_data['city_3'])
```

```python
city_names = ['City_A', 'City_B', 'City_C', 'City_D', 'City_E', 'City_F', 'City_G', 'City_H', 'City_I', 'City_J']

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

city_df = pd.DataFrame(temperature, index=city_names, columns= months)
city_df
```

#Data Analysis

Calculate the annual average temperature for each city.

Identify the city with the highest and lowest average temperature for the year.

```python
average_temp = city_df.mean(1)
average_temp
```

```python
hotest_city = average_temp.idxmax()
print(f"The hotest city is {hotest_city}")
coldest_city = average_temp.idxmin()
print(f"The coldest city is {coldest_city}")
```

# Data Visualization

```python
import matplotlib.pyplot as plt

# Create subplots for each city
num_cities = len(city_df)
num_rows = (num_cities + 1) // 2  # Ensures at least 2 rows
fig, axes = plt.subplots(num_rows, 2, figsize=(20, 5*num_rows))
axes = axes.flatten()

# Plot temperature for each city in each subplot
for i, city in enumerate(city_df.index):
    ax = axes[i]
    ax.plot(city_df.columns, city_df.loc[city], label=city, color='skyblue')
    ax.set_title(f'Temperature in {city}')
    ax.set_xlabel('Month')
    ax.set_ylabel('Temperature')
    ax.grid(True)


# Remove unused subplots
for j in range(num_cities, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Transpose the DataFrame to have months as rows and cities as columns
transposed_df = city_df.transpose()

# Plot temperature for each city across all months
plt.figure(figsize=(12, 8))
for city in transposed_df.columns:
    plt.plot(transposed_df.index, transposed_df[city], label=city)

plt.title('Temperature for each city across all months')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Transpose the DataFrame to have months as rows and cities as columns
transposed_df = city_df.transpose()

# Plot temperature for each city across all months
plt.figure(figsize=(12, 8))
months = range(1, 13)
bottom = None  # Variable to keep track of bottom values for stacking bars
for city in transposed_df.columns:
    plt.bar(months, transposed_df[city], label=city, bottom=bottom)
    if bottom is None:
        bottom = transposed_df[city]
    else:
        bottom += transposed_df[city]

plt.title('Sum of weather along the year')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.xticks(months, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
