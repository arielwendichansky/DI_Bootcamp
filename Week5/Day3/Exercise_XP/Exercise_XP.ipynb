```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import math
from sklearn.metrics import roc_curve, roc_auc_score
from scipy import stats, cluster
from scipy.optimize import minimize_scalar
from scipy.stats import skew, kurtosis, median_test,ttest_ind, norm
from plotnine import (ggplot, aes, geom_point, theme_matplotlib, theme_set,labs, ggtitle,geom_col,geom_line)
```

# ðŸŒŸ Exercise 1 : Basic Scatter Plot With Plotnine

```python
np.random.seed(42)
array1 = np.random.randint(0,20,size=20)
array2 = np.random.randint(0,20,size=20)

data = {'X':array1,
        'Y':array2}
df = pd.DataFrame(data)
df
```

```python
gg = ggplot(df, aes(x='X', y='Y')) + geom_point() + theme_matplotlib()+ labs(x='X-Value', y='Y-Value') + ggtitle('Basic Scatter Plot')

print(gg)
```

# ðŸŒŸ Exercise 2 : Bar Plot Of Categorical Data With Plotnine

```python
np.random.seed(42)

category = np.random.choice(['A','B','C'], size=30)
array1 = np.random.randint(0,21, size=30)


data = {'category':category,
        'values':array1}

df = pd.DataFrame(data)
df
```

```python
gg = ggplot(df, aes(x="category", y="values", fill="category")) + geom_col()+labs(x='x-axis', y='y-axis' )+ggtitle('Category Frequency')
print(gg)
```

# ðŸŒŸ Exercise 3 : Line Plot With Multiple Lines With Plotnine

```python
np.random.seed(42)
x = np.random.randint(0,21,size=30)

y1=np.sin(x)
y2=np.cos(x)

data = {'X':x,
        'Y1':y1,
        'Y2':y2}

df = pd.DataFrame(data)
df
```

```python
gg = ggplot(df) + aes(x='x') + geom_line(aes(y='y1'), color='blue') + geom_line(aes(y='y2'), color='red')+labs(x='x-axis', y='y-axis' )+ggtitle('Comparison of Two Functions')

print(gg)
```

 # Exercise 4 : Basic Line Chart With Plotly

```python
data = {'months' : ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
'sales' : [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]}

df=pd.DataFrame(data)
df
```

# Exercise 5 : Bar Chart With Customization With Plotly

```python
data= {'genres':  ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography'],
'books_sold': [120, 150, 90, 80, 130]}
df=pd.DataFrame(data)
df
```
