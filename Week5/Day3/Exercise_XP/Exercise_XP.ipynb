# ðŸŒŸ Exercise 2 : Descriptive Statistics
Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
Sample dataset in code

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]

avg_sp = stats.tmean(data)
print('The mean is:', avg_sp)
median_np = np.median(data)
print('The median is:', median_np)
variance_sp = stats.tvar(data)
print('The variance is:', variance_sp)
std_dev_sp = stats.tstd(data)
print('The standard deviation is:', std_dev_sp)
```

#ðŸŒŸ Exercise 3 : Understanding Distributions
Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.



```python
mean = 50
standard_deviation = 10

# Create normal distribution using the 'norm' class
# loc - specifies the mean
# scale - specifies the standard deviation
data = stats.norm( loc=mean, scale=standard_deviation)

sample = data.rvs(size=4000)
sample.round(2)

### Plot sample heights as histogram

# Use seaborn style and increase font size
sns.set_theme(style='ticks', font_scale=1.5)
# set figure dimensions
plt.figure(figsize=(12, 6))

# Plot histogram
# Use larger bin count (100) to get smoother shape
plt.hist(sample, bins=100)

plt.title("Random distribution - Histogram", pad=20)
plt.xlabel("Value", labelpad=10)
plt.ylabel("Frequency", labelpad=10)
plt.show()
```

# ðŸŒŸ Exercise 4 : T-Test Application

Perform a T-test on two sets of randomly generated data. Hint: Use numpy.random.normal() to generate the datasets.

```python
test_1 = np.random.normal(0.0, 1.0, 50)
test_2 = np.random.normal(0.0, 1.0, 50)

t_statistic, p_value = stats.ttest_ind(test_1, test_2)

if p_value <= 0.05 and p_value >= - 0.05:
  print("Reject the null hypothesis")
  print("t-statistic:", t_statistic)
  print("p-value:", p_value)
else:
  print("Accept the null hypothesis")
  print("t-statistic:", t_statistic)
  print("p-value:", p_value)
```

# ðŸŒŸ Exercise 6 : Advanced Statistical Functions

Explore one advanced statistical function in SciPy and demonstrate its use with an example.

Sample Task: Calculate kurtosis and skewness of a dataset.


```python
# Graph using numpy.linspace()
# finding kurtosis

x1 = np.linspace( -5, 5, 1000 )
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  )

p.plot(x1, y1, '*')


print( '\nKurtosis for normal distribution :', kurtosis(y1))

print( '\nKurtosis for normal distribution :',
      kurtosis(y1, fisher = False))

print( '\nKurtosis for normal distribution :',
      kurtosis(y1, fisher = True))
```

```python
# finding Skewness

x1 = np.linspace( -5, 12, 1000 )
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  )

p.plot(x1, y1, '.')

print( '\nSkewness for data : ', skew(y1))
```
