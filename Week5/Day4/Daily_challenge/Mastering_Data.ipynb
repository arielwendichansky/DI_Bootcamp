```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import math
from sklearn.metrics import roc_curve, roc_auc_score
from scipy import stats, cluster
from scipy.optimize import minimize_scalar
from scipy.stats import skew, kurtosis, median_test,ttest_ind, norm
from plotnine import (ggplot, aes, geom_point, theme_matplotlib, theme_set,labs, ggtitle,geom_col,geom_line)
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.describe().T
```

```python
plt.figure(figsize=(15,6))

plt.hist(data=df, x='petal_length')
plt.title('Petal Lengths Distribution')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set_theme(style="ticks")
fig = sns.pairplot(df, hue='species')

plt.suptitle('Pairwise Relationships',  y=1.05)
plt.show()
```

```python
gg = (
    ggplot(df, aes("sepal_length", "sepal_width"))
    + geom_point(aes(color="species"))
    + ggtitle('Correlation Sepal and Species')
)

print(gg)
```

```python
fig = px.scatter_3d(df, x= 'sepal_length', y='sepal_width',z='petal_length', color='species')
fig.show()
```
